# Dataset Preview
Use this notebook to inspect a handful of samples from the chosen code dataset before launching RLOO training.

```python
# Optional: install project requirements if running outside Kaggle dataset mount
# %pip install -r ../requirements.txt

import sys
from pathlib import Path

# point to the uploaded dataset path on Kaggle
DATASET_ROOT = Path('/kaggle/input/rloo')
SRC_PATH = DATASET_ROOT / 'src'
if SRC_PATH.exists():
    sys.path.append(str(SRC_PATH))

from rloo_project.config import TrainingConfig
from rloo_project.data import get_source_column, load_code_dataset

config = TrainingConfig.from_yaml(DATASET_ROOT / 'configs' / 'default_training.yaml')
dataset = load_code_dataset(config.dataset)
column = get_source_column(dataset)
column
```

```python
import itertools
from pprint import pprint

# Display a small sample
for row in itertools.islice(dataset, 3):
    pprint({column: row[column][:200]})
    print('-' * 40)
```
